```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```python
print(X)
```

```python
print(y)
```

Splitting the dataset into training and testing part

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Model Training

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Predicting the test result

```python
y_pred = regressor.predict(X_test)
```

Making a single prediction (for ex - the salary of an employee with 12 years of experience)

```python
print(regressor.predict([[12]]))
```

Note - Notice that the value of the feature (12years) was input in a double pair of square braackets. That's because the "predict" method always excepts a 2D array as the format of its input. and putting 12 into a double pair of square brackets makes the imput exactly a 2D array.

Visulaizing the training set result

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
