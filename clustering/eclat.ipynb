```python
import pandas as pd
from collections import defaultdict
from itertools import combinations
```

```python
df = pd.read_csv('Groceries_dataset.csv')
```

```python
#Group items by transaction (same customer & date = 1 transaction)
df['itemDescription'] = df['itemDescription'].astype(str)
grouped = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(list)
transactions = grouped.tolist()
transactions
```

```python
vertical_db = defaultdict(set)
for tid, items in enumerate(transactions): #enumerate to get transaction id
    for item in items:
        vertical_db[item].add(tid)
```

```python
for items, tids in vertical_db.items():
    print(f"Item: {items} => Transaction IDs: {sorted(list(tids))}")
```

```python
min_support = int(0.002 * len(transactions))
```

```python
frequent_itemsets = {}
for items, tids in vertical_db.items():
    if len(tids) >= min_support:
        frequent_itemsets[items] = tids
```
